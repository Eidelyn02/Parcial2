```python
import math as ma 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
r=x**2
pi=3.14
y=3.0*r**2*np.exp(-1.5*x)*np.sin(2*x*np.pi)

plt.plot(x,y,"-",color="GREEN")
plt.xlabel("x")
plt.ylabel("y")
plt.title("funcion f(X)=3.0*r2*exp(-1.5*x)*sin(2*x*PI)")
plt.grid()

plt.show()
```

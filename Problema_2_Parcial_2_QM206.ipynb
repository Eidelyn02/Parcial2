```python
url="https://docs.google.com/spreadsheets/d/1WUZstMoBDohgOIxHd7V2UzlMIaqhVoBVpLev02aeAMA/export?format=csv"

df=pd.read_csv(url)
df
```

```python
PFs=df["PF"]
PEb=df["PE"]
NA=df["NA"]


plt.plot(NA,PFs, color="green")
plt.plot(NA,PEb, color="blue")
plt.xlabel("Punto de ebullicion (azul) y punto de fusion(verde)")
plt.ylabel("Numero Atomico (g/mol)")
plt.title("Punto de Fusion(PF) y Punto de Ebullicion(PE) VS Numero Atomico(NA)")
plt.show()
```
